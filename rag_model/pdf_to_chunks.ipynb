```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
import json
```

```python
# load the pdf

loader = PyPDFLoader("notes/dsa_book.pdf")
pages = loader.load()
```

```python
# split the pdf into chunks

splitter = RecursiveCharacterTextSplitter(
    chunk_size=600,
    chunk_overlap=100,
    length_function=len,
    separators=["\n\n", "\n", " ", ""],
)

chunks = splitter.split_documents(pages)
```

```python
# convert document objects to JSON format
json_chunks = [
    {"content": chunk.page_content, "metadata": chunk.metadata} for chunk in chunks
]
```

```python
# save chunks to chunks.json

with open("json/chunks.json", "w", encoding="utf-8") as f:
    json.dump(json_chunks, f, indent=4, ensure_ascii=False)

print(f"Saved {len(json_chunks)} chunks to chunks.json")
```
